```python
import pandas as pd
df = pd.read_csv(r"C:\Users\Asus\Desktop\Data cleaning file .csv")
```

```python
df.head(15)
```

```python
# Remove bracket part
df["Client"] = df["Client"].str.split("(").str[0]

# Remove extra spaces
df["Client"] = df["Client"].str.strip()

# Proper formatting
df["Client"] = df["Client"].str.title()

# Clean contact names
df["Contact"] = (
    df["Contact"]
    .str.strip()
    .str.replace(r"\s+", " ", regex=True)
    .str.title()
)

# Split department and city
df[["Department", "City"]] = df["Department"].str.split("_", expand=True)

# Fill null values
df = df.fillna("NA")

# Remove duplicates
df = df.drop_duplicates()

# Replace #DIV/0!
df["Profit Margin"] = df["Profit Margin"].replace("#DIV/0!", "Null")
```

```python
df = df[[
    "Date",
    "Client",
    "Contact",
    "Department",
    "City",
    "Payment",
    "Revenue",
    "Profit",
    "Profit Margin"
]]
```

```python
df.to_csv("cleaned_data.csv", index=False)
```
